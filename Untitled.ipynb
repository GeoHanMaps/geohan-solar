```python
# Konya-Ankara bölgesi
bbox = ee.Geometry.BBox(31.999, 37.999, 33.000, 39.000)

# Global Solar Atlas - güneş ışınımı verisi
ghi = ee.ImageCollection("projects/sat-io/open-datasets/global-solar-atlas/ghi") \
    .first() \
    .clip(bbox)

# İstatistikleri hesapla
stats = ghi.reduceRegion(
    reducer=ee.Reducer.mean().combine(
        ee.Reducer.minMax(), sharedInputs=True
    ),
    geometry=bbox,
    scale=250,
    maxPixels=1e8,
).getInfo()

print("Ortalama GHI:", round(stats['b1_mean'], 2), "kWh/m²/yıl")
print("Min GHI:", round(stats['b1_min'], 2))
print("Max GHI:", round(stats['b1_max'], 2))
```

```python
import requests

# NASA POWER API - güneş ışınımı verisi
# Bölgenin merkez noktası
lat = 38.5
lon = 32.5

url = f"https://power.larc.nasa.gov/api/temporal/climatology/point?parameters=ALLSKY_SFC_SW_DWN&community=RE&longitude={lon}&latitude={lat}&format=JSON"

response = requests.get(url)
veri = response.json()

ghi_aylik = veri['properties']['parameter']['ALLSKY_SFC_SW_DWN']

print("Aylık GHI (kWh/m²/gün):")
for ay, deger in ghi_aylik.items():
    print(f"  {ay}: {deger}")
```

```python
# Yıllık toplam hesapla
yillik_ghi = ghi_aylik['ANN'] * 365

print(f"Yıllık toplam GHI: {yillik_ghi:.0f} kWh/m²/yıl")

# GES üretim tahmini (basit hesap)
# 1 MW GES için yaklaşık 2 hektar arazi
# Performans oranı: 0.80
uretim_1mw = yillik_ghi * 0.80
print(f"1 MW GES yıllık üretim tahmini: {uretim_1mw:.0f} kWh/yıl")
print(f"1 MW GES yıllık üretim tahmini: {uretim_1mw/1000:.0f} MWh/yıl")
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

veri = rasterio.open("n38_e032_1arc_v3.tif")
yukseklik = veri.read(1)
print("Veri yüklendi!")
```

```python
piksel_metre = 30
x_egim = np.gradient(yukseklik, piksel_metre, axis=1)
y_egim = np.gradient(yukseklik, piksel_metre, axis=0)
egim = np.sqrt(x_egim**2 + y_egim**2)
egim_derece = np.degrees(np.arctan(egim))
print("Ortalama eğim:", egim_derece.mean().round(2), "derece")
```
